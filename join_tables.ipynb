# Join Tables

```python
# This script joins the servespeed.csv with player metrics
```

```python
# import libraries
import pandas as pd
```

```python
servespeed = pd.read_csv('servespeed.csv')
metrics = pd.read_csv('tennis_players.csv')

# reverse the order of names in tennis_players.csv to match the format in servespeed.csv
metrics['name'] = metrics['name'].apply(lambda x: ' '.join(x.split(' ')[::-1]))

# merge the two dataframes on the 'name' column
merged = pd.merge(servespeed, metrics, on='name')
```
